## Task 1
Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df_heart=pd.read_csv('Heart.csv')
```

```python
gender={1:"Male",0:"Female"}
df_heart['sex'].replace([1,0],['Male','Female'],inplace=True)
```

```python
df_heart
```

```python
bins = [29, 40, 50, 60, 77]
labels = ['29-39', '40-49', '50-59', '60-77']
df_heart['Age Range'] = pd.cut(df_heart['age'].astype(float), bins=bins, labels=labels, right=False)
```

```python
df_heart.info()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].set_title("Histogram for distribution of Age")
axs[0].hist(df_heart["Age Range"].dropna().sort_index(), bins=20, color="skyblue")
for age, count in df_heart["Age Range"].value_counts().items():
    axs[0].text(age, count, str(count), ha='center', va='bottom')
axs[0].set_xlabel("Age Range")
axs[0].set_ylabel("Frequency")
axs[1].hist(df_heart["sex"], color="salmon", bins=3)
axs[1].set_title("Histogram for Distribution of Gender")
axs[1].set_xlabel("Gender")
axs[1].set_ylabel("Frequency")
axs[1].grid(True)
for gender, count in df_heart["sex"].value_counts().items():
    axs[1].text(gender, count, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid") 
plt.figure(figsize=(10, 6))  
sns.distplot(df_heart['age'], kde=True, color='skyblue', hist_kws={'edgecolor': 'black'}) 
plt.xlabel('Age', fontsize=14)  
plt.ylabel('Density', fontsize=14)  
plt.title('Age Distribution', fontsize=16) 
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)  
plt.tight_layout()  
plt.show()
```

```python
sns.set(style="whitegrid")  
plt.figure(figsize=(10, 6)) 
sns.histplot(data=df_heart, x='sex', discrete=True, kde=False, color='skyblue') 
plt.xlabel('Sex', fontsize=14)  
plt.ylabel('Frequency', fontsize=14)  
plt.title('Sex Distribution', fontsize=16) 
plt.xticks(fontsize=12) 
plt.yticks(fontsize=12) 
plt.tight_layout()  
plt.show()
```

```python
sns.set(style="whitegrid")  
plt.figure(figsize=(10, 6)) 
sns.countplot(data=df_heart, x='sex',color='skyblue') 
plt.xlabel('Sex', fontsize=14)  
plt.ylabel('Count', fontsize=14)  
plt.title('Sex Count', fontsize=16) 
plt.xticks(fontsize=12) 
plt.yticks(fontsize=12) 
plt.tight_layout()  
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot Age Range vs trtbps
sns.barplot(data=df_heart, x='Age Range', y='trtbps', ax=axs[0, 0], palette='coolwarm')
axs[0, 0].set_title('Age Range vs trtbps')
axs[0, 0].set_xlabel('Age Range')
axs[0, 0].set_ylabel('trtbps')

# Plot Age Range vs chol
sns.barplot(data=df_heart, x='Age Range', y='chol', ax=axs[0, 1], palette='husl')
axs[0, 1].set_title('Age Range vs chol')
axs[0, 1].set_xlabel('Age Range')
axs[0, 1].set_ylabel('chol')

# Plot sex vs trtbps
sns.barplot(data=df_heart, x='sex', y='trtbps', ax=axs[1, 0], palette='Set2')
axs[1, 0].set_title('sex vs trtbps')
axs[1, 0].set_xlabel('sex')
axs[1, 0].set_ylabel('trtbps')

# Plot sex vs chol
sns.barplot(data=df_heart, x='sex', y='chol', ax=axs[1, 1], palette='tab10')
axs[1, 1].set_title('sex vs chol')
axs[1, 1].set_xlabel('sex')
axs[1, 1].set_ylabel('chol')

plt.tight_layout()
plt.show()
```
